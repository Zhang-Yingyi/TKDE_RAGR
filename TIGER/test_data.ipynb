```python
save_jsonl = args.data_path+"/{}/{}_train_results.jsonl".format(args.dataset,args.dataset)
data = []
with open(save_jsonl, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        tmp_data = json.loads(line)
        for k,v in tmp_data.items():
            if type(v) == str:
                tmp_data[k] = v.replace(" ", "")
            else:
                tmp_data[k] = [vi.replace(" ", "") for vi in v]
        data.append(tmp_data)

print("loaded:", len(data))
print("first sample:", data[0])
```
